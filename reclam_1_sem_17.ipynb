### Assuntos listados - nem todos. Abaixo está a lista pelo número de vezes que aparecem, ou seja, assuntos mais reclamados - queixas sobre Banco de Dados e Cadastros de Consumidores (SPC, Serasa, SCPC etc) lideram 

```python
reclamacoes.Assunto
```

```python
reclamacoes.Assunto.value_counts().reset_index()
```

### E os principais problemas encontrados - Dados pessoais ou financeiros consultados, coletados, publicados ou repassados sem autorização, está na liderança

### Mostra as cidades com mais esse tipo de queixa
